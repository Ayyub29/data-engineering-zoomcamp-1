```python
import pandas as pd
import pandas.io.sql
from time import time
pd.__version__
```

```python
from sqlalchemy import create_engine

engine = create_engine(f"postgresql://postgres:root123@localhost:5431/ny_taxi")
conn = engine.connect()
```

```python
df_ny_taxi = pd.read_csv("yellow_tripdata_2021-01.csv", iterator=True, chunksize=100000)

df_cur = next(df_ny_taxi)
df_cur.head(10)
```

```python
len(df_cur)
```

```python
df_cur.tpep_pickup_datetime = pd.to_datetime(df_cur.tpep_pickup_datetime)
df_cur.tpep_dropoff_datetime = pd.to_datetime(df_cur.tpep_dropoff_datetime)

df_cur.head(10)
```

```python
# upload to postgres

%time df_cur.head(n=0).to_sql(name='yellow_taxi', con=conn, if_exists='replace')
```

```python
%time df_cur.to_sql(name='yellow_taxi', con=conn, if_exists='append')
```

```python
while True: 
    t_start = time()

    df_cur = next(df_ny_taxi)

    df_cur.tpep_pickup_datetime = pd.to_datetime(df_cur.tpep_pickup_datetime)
    df_cur.tpep_dropoff_datetime = pd.to_datetime(df_cur.tpep_dropoff_datetime)

    df_cur.to_sql(name='yellow_taxi', con=conn, if_exists='append', method='multi')

    t_end = time()

    print('inserted another chunk, took %.3f second' % (t_end - t_start))
```

```python
#How many taxi trips were there on January 15?

query = "SELECT count(1) FROM yellow_taxi WHERE tpep_pickup_datetime BETWEEN '2021-01-15 00:00:00.00' AND '2021-01-15 23:59:59.999';"
pd.read_sql(query, con=conn)
```

```python
#On which day it was the largest tip in January? (note: it's not a typo, it's "tip", not "trip")

query = "SELECT CAST(tpep_pickup_datetime AS DATE), MAX(tip_amount) FROM yellow_taxi WHERE EXTRACT(MONTH FROM tpep_pickup_datetime)=1 GROUP BY CAST(tpep_pickup_datetime AS DATE) ORDER BY max DESC LIMIT 3;"

pd.read_sql(query, con=conn)
```

```python
#On which day it was the largest tip in January? (note: it's not a typo, it's "tip", not "trip")

query = "SELECT CAST(tpep_pickup_datetime AS DATE), SUM(tip_amount) FROM yellow_taxi GROUP BY CAST(tpep_pickup_datetime AS DATE) ORDER BY sum DESC LIMIT 3;"
pd.read_sql(query, con=conn)
```

```python
# What was the most popular destination for passengers picked up in central park on January 14? Enter the zone name (not id). If the zone name is unknown (missing), write "Unknown"

query = """SELECT "Zone", count("Zone") FROM yellow_taxi LEFT JOIN zones ON zones."LocationID"=yellow_taxi."PULocationID" GROUP BY "Zone" ORDER BY count DESC LIMIT 3;"""
pd.read_sql(query, con=conn)
```

```python
#What's the pickup-dropoff pair with the largest average price for a ride (calculated based on total_amount)? 
# Enter two zone names separated by a slashFor example:"Jamaica Bay / Clinton East"
#If any of the zone names are unknown (missing), write "Unknown". For example, "Unknown / Clinton East".

query = """SELECT zones_pickup."Zone" AS Zone_PU,zones_dropoff."Zone" AS Zone_DO, AVG(total_amount) FROM yellow_taxi LEFT JOIN zones zones_pickup ON zones_pickup."LocationID"=yellow_taxi."PULocationID" LEFT JOIN zones zones_dropoff ON zones_dropoff."LocationID"=yellow_taxi."DOLocationID" GROUP BY Zone_PU, Zone_DO ORDER BY avg DESC LIMIT 3;"""
pd.read_sql(query, con=conn)
```
